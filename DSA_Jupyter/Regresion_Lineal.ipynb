```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dsets = pd.read_csv('Boston House Prices.csv')
dsets
```

```python
y = dsets['Value']
x = dsets['Rooms']
```

```python
dsets.plot(kind='scatter', x='Rooms', y='Value')
```

```python
x=sm.add_constant(x)
linear_model= sm.OLS(y,x).fit()
```

```python
linear_model.predict()
```

```python
linear_model.summary()
```

```python
y_pred=9.1021*x['Rooms']-34.6706
```

```python
#GRAFICAR LOS PUNTOS
plt.figure(figsize=(6,4), tight_layout=True)
sns.scatterplot(x=x['Rooms'], y=y)

#GRAFICAR LA LINEA DE DISPERCION
sns.lineplot(x=x['Rooms'], y=y_pred, color='red')

#AXIS
plt.xlim(0)
plt.ylim(0)
plt.savefig('Linear_Regression')
plt.show()
```
